```python
df1["gender"] =df1["gender"].map({"Male":0 ,"Female":1})
```

```python
df1["cough"] =df1["gender"].map({"Mild":0 ,"Strong":1})
```

```python
df1["city"] =df1["city"].map({"Kolkata":0 ,"Bangalore":1 ,"Dehli" :2 ,"Mumbai":3})
```

```python
df1["has_covid"] =df1["has_covid"].map({"No":0 ,"Yes":1})
```

```python
df1.head(3)
```

```python
df1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si =SimpleImputer(strategy ="mean")
```

```python
df1["fever"] =si.fit_transform(df1[["fever"]]) 
df1["cough"] =si.fit_transform(df1[["cough"]])
df1["city"] =si.fit_transform(df1[["city"]])#fit means learns  the perametre and transform
```
